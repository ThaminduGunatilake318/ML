````python
import csv
import requests
from io import StringIO
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

file_path = '/content/drive/MyDrive/ML/train.csv'

data_set = pd.read_csv(file_path)
print("Data_set shape after read:", data_set.shape)
# data_set.dropna(inplace=True)  # Remove rows with NaN values
# print("Data_set shape after remioving Nan:", data_set.shape)
data_set.drop_duplicates(inplace=True) #removes duplicates
print("Data_set shape after removing duplicates:", data_set.shape)


print(" values in numerical columns before imputation:")
print(data_set.isna().sum())
print("Data_set shape after///:", data_set.shape)

data_set.dropna(subset=['ID'], inplace=True) # dropping rows of empty ID's

print("lues in numerical columns before imputation:")
print(data_set.isna().sum())
print("Data_set shape after readafdad:", data_set.shape)



# Select only the numerical columns for imputation
numerical_columns = data_set.select_dtypes(include=[np.number]).columns
# Perform imputation
# Check for NaN values in numerical columns after imputation
print("NaN values in numerical columns before imputation:")
print(data_set[numerical_columns].isna().sum())

# Create an instance of SimpleImputer with the desired strategy (e.g., 'mean', 'median', 'most_frequent')
imputer = SimpleImputer(strategy='median')  # You can change the strategy as needed
# Fit the imputer to the selected numerical columns
imputer.fit(data_set[numerical_columns])
# Transform the selected numerical columns to replace incorrect values with imputed values
data_set[numerical_columns] = imputer.transform(data_set[numerical_columns])
print("Data_set shape after imputation:", data_set.shape)
print(data_set.head(62).tail(65))
print("NaN values in numerical columns after imputation:")
print(data_set[numerical_columns].isna().sum())




#outliers detection
z_score_threshold = 3
# Filter rows where z-scores are within threshold for all numeric columns
data_set_no_outliers = data_set[(np.abs(data_set.select_dtypes(include=[np.number]).apply(
    lambda x: (x - x.mean()) / x.std())) < z_score_threshold).all(axis=1)]
print("Original data_set shape:", data_set.shape)
print("Data_set shape after removing outliers:", data_set_no_outliers.shape)
# print(data_set_no_outliers.head(12).tail(5))


# # Select only the numerical columns for imputation
# numerical_columns = data_set_no_outliers.select_dtypes(include=[np.number]).columns
# # Create an instance of SimpleImputer with the desired strategy (e.g., 'mean', 'median', 'most_frequent')
# imputer = SimpleImputer(strategy='mean')  # You can change the strategy as needed
# # Fit the imputer to the selected numerical columns
# imputer.fit(data_set_no_outliers[numerical_columns])
# # Transform the selected numerical columns to replace incorrect values with imputed values
# data_set_no_outliers[numerical_columns] = imputer.transform(data_set_no_outliers[numerical_columns])



print("Original data_set shape:", data_set.shape)
print("Data_set shape after removing incorrect values:", data_set_no_outliers.shape)

# xw = data_set_no_outliers.iloc[:, :-1].values #fetching all rows anc cols except target

#making 2 matrices for train and test
x = data_set.iloc[:, :-1].values #fetching all rows anc cols except target
y = data_set.iloc[:, -1].values # fetching all rows and the last col





# num_rows_x = x.shape[0]  # Get the number of rows in x
# print("Number of rows in x:", num_rows_x)
# num_col_x = x.shape[1]  # Get the number of rows in x
# print("Number of col in x:", num_col_x)

# num_rows_y = y.shape[0]  # Get the number of rows in x
# print("Number of rows in y:", num_rows_y)
````
